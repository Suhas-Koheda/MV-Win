```python
import cv2
import urllib.request
import numpy as np
import matplotlib.pyplot as plt
```

```python
req = urllib.request.urlopen('https://d3t4nwcgmfrp9x.cloudfront.net/upload/nuevas-herramientas-gmail.png')
arr = np.asarray(bytearray(req.read()), dtype=np.uint8)
img = cv2.imdecode(arr, -1)
print(img.shape)
img
```

```python
small=cv2.resize(img,None,fx=0.3,fy=0.3,interpolation=cv2.INTER_AREA)
print(small.shape)
small
```

| Method            | When to Use       | Description                               |
|-------------------|-------------------|-------------------------------------------|
| `cv2.INTER_AREA`  | Shrinking         | Minimizes distortion while downscaling.   |
| `cv2.INTER_LINEAR`| General resizing  | Balances speed and quality.               |
| `cv2.INTER_CUBIC` | Enlarging         | Higher quality for upscaling.             |
| `cv2.INTER_NEAREST`| Fast resizing    | Quick but lower quality.                  |

```python
plt.figure(figsize=(10,10))
# plt.imshow(img) # shows the brg image directly
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB)) # converts BGR to RGB for matplotlib
plt.show()
```

```python
rotate=cv2.rotate(img,cv2.ROTATE_90_CLOCKWISE)
plt.imshow(cv2.cvtColor(rotate,cv2.COLOR_BGR2RGB))
```

```python
rotate=cv2.rotate(img,cv2.ROTATE_90_COUNTERCLOCKWISE)
plt.imshow(cv2.cvtColor(rotate,cv2.COLOR_BGR2RGB))
```

```python
height,width=img.shape[:2]
quar_h,quar_w=height/2,width/2
T=np.float32([[1,0,quar_w],[0,1,quar_h]])
translated=cv2.warpAffine(img,T,(width,height)) ## fucntion required to translate image
# translation performs the operation using [[ 1 0 Tx][ 0 1 Ty]]
# Tx is the horizontal shift
# Ty is the vertical shift
translated
```

```python
def rotate_image(image, angle):
  image_center = tuple(np.array(image.shape[1::-1]) / 2)
  rot_mat = cv2.getRotationMatrix2D(image_center, angle, 1.0)
  print(rot_mat)
  result = cv2.warpAffine(image, rot_mat, image.shape[1::-1], flags=cv2.INTER_LINEAR)
  return result
rotate_image(img,30) # degree anticlockwise
```
